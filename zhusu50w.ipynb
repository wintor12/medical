```python
import jieba
import gensim, logging
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
```

```python
with open('/home/tongwang/data/medical/13') as p:
    s = p.read()
lines = s.split('\n')
with open('/home/tongwang/data/medical/14') as p:
    s2 = p.read()
lines += s2.split('\n')
```

```python
import re
aa = lines[0]
print(aa)
bb = re.sub(r'[a-zA-Z\d\(\)\-\,\/\s+\.]','',aa)
# bb = re.sub('[\s+\.\!\/_,$%^*(+\"\'[a-zA-Z]\d]+|[+——！，。？、~@#￥%……&*（）]+'.decode('utf8'), '',aa)  
print(bb)
cc = list(jieba.cut(bb, cut_all=False))
for x in cc:
    print(x)
```

```python
print(len(lines))
lines = [re.sub(r'[a-zA-Z\d\(\)\-\,\/\s+\.]','',l) for l in lines]
print(lines[0])
```

```python
sentences = []
for i in range(len(lines)):
    if i%1000 == 0:
        print(i)
    sentences.append(list(jieba.cut(lines[i], cut_all=False)))
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
model = gensim.models.Word2Vec(sentences, min_count=3)
```

```python
words = model.most_similar(u'头痛')
for word in words:
    print word[0],word[1]
```

```python
vocab = model.vocab.keys()
res = []
for x in vocab:
    words = model.most_similar(x)
    words = [y[0] for y in words]
    res.append(x + ',' + ','.join(words))

with open('vocab.txt', 'w') as p:
    p.write('\n'.join(res).encode('utf-8'))
```
